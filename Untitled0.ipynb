```python
df.info()
```

```python
from bs4 import BeautifulSoup
def remove_html(text):
  soup=BeautifulSoup(text, 'lxml')
  html_free=soup.get_text()
  return html_free

df['content']= df['content'].apply(lambda x: remove_html(x))
```

```python
def cleaning_numbers(data):
    return re.sub('[0-9]+','', data)
df['content']= df['content'].apply(lambda x: cleaning_numbers(x))
```

```python
from nltk import WhitespaceTokenizer
words=[]
tokenizer=WhitespaceTokenizer()
words=df.content.apply(tokenizer.tokenize)
words
```

```python
df['words']=words
df
```

```python
def text_to_lower(text):
  txt=[word.lower() for word in text]
  return txt

df['words'] = df['words'].apply(lambda x: text_to_lower(x)) 
```

```python
st = nltk.PorterStemmer()
def stemming_on_text(data):
    data = [st.stem(word) for word in data]
    return data
df['words']= df['words'].apply(lambda x: stemming_on_text(x))
```

```python
freq_dist=[FreqDist(word) for word in df.words]
print(freq_dist)
```
